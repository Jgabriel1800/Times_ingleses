```python
import pandas as pd
import numpy as np
import matplotlib as plt
import plotly.express as px
```

```python
jogos_fora = dados_inicio.groupby('AwayTeam')['AwayTeam'].transform('count')
```

```python
times = pd.concat([dados_inicio['HomeTeam'], dados_inicio['AwayTeam']]).unique()
times=pd.DataFrame(times,columns=['times'])
```

**Criando colunas**

Criando coluna de times

```python
dados_inicio['Times']=times
dados_inicio['Times'].sort_values()
dados_inicio['Times'] = dados_inicio['Times'].sort_values().values
dados_inicio
```

Criando coluna de gols dos times

```python
gols_time=dados_inicio.groupby('HomeTeam')['HomeGoals'].sum() +dados_inicio.groupby('AwayTeam')['AwayGoals'].sum()
dados_inicio['Gols time'] = dados_inicio['Times'].map(gols_time).sort_values().fillna(0)
dados_inicio
```

Criando coluna gols dentro e fora de casa

```python
gols_time_casa=dados_inicio.groupby('HomeTeam')['HomeGoals'].sum()
dados_inicio['Gols casa']=dados_inicio['Times'].map(gols_time_casa).sort_values().fillna(0)
dados_inicio
```

```python
gols_fora=dados_inicio.groupby('AwayTeam')['AwayGoals'].sum()
dados_inicio['Gols fora']=dados_inicio['Times'].map(gols_fora).sort_values().fillna(0)
dados_inicio
```

```python
todos_jogos=dados_inicio.groupby('HomeTeam')['HomeTeam'].count() + dados_inicio.groupby('AwayTeam')['AwayTeam'].count()
dados_inicio['Total jogos']=dados_inicio['Times'].map(todos_jogos).sort_values().fillna(0)
dados_inicio
```

```python
todos_jogos_fora=dados_inicio.groupby('HomeTeam')['HomeTeam'].count()
todos_jogos_casa=dados_inicio.groupby('AwayTeam')['AwayTeam'].count()
dados_inicio['Jogos Casa']=dados_inicio['Times'].map(todos_jogos_casa).sort_values().fillna(0)
dados_inicio['Jogos Fora']=dados_inicio['Times'].map(todos_jogos_fora).sort_values().fillna(0)
dados_inicio
```

Modificando os tipos das colunas

```python
dados_inicio[['Gols time','Gols casa','Gols fora','Total jogos','Jogos Fora','Jogos Casa','HomeGoals','AwayGoals']]=dados_inicio[['Gols time','Gols casa','Gols fora','Total jogos','Jogos Fora','Jogos Casa','HomeGoals','AwayGoals']].astype(np.int64)
dados_inicio.info()
dados_inicio
```

removendo coluna

```python
dados_inicio=dados_inicio.drop(columns=['League'])
dados_inicio
```

```python
dados_inicio['Media gols']=(dados_inicio['Gols time']/dados_inicio['Total jogos']).fillna(0)
dados_inicio['Media gols casa']=(dados_inicio['Gols casa']/dados_inicio['Total jogos']).fillna(0)
dados_inicio['Media gols fora']=(dados_inicio['Gols fora']/dados_inicio['Total jogos']).fillna(0)

dados_inicio
```
